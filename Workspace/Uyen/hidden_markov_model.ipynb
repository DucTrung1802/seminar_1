```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from hmmlearn.hmm import GaussianHMM
from datetime import timedelta
from sklearn.preprocessing import StandardScaler

# ---------------------------
# 1. Data Loading and Preprocessing
# ---------------------------
# Replace 'your_dataset.csv' with your actual file path.
# The CSV should contain at least the columns: "date" and "revenue".
df = pd.read_csv("df.csv")
df = df.fillna(0)
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date').reset_index(drop=True)

# ---------------------------
# 2. Feature Engineering
# ---------------------------
# Basic time features
df['dayofweek'] = df['date'].dt.dayofweek           # Monday=0, Sunday=6
df['is_weekend'] = df['dayofweek'].isin([5, 6]).astype(int)
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
df['day_of_year'] = df['date'].dt.dayofyear

# Use sine and cosine transforms for day-of-week to capture cyclical behavior
df['sin_day'] = np.sin(2 * np.pi * df['dayofweek'] / 7)
df['cos_day'] = np.cos(2 * np.pi * df['dayofweek'] / 7)

# Lag feature: revenue from the previous day
df['lag1'] = df['revenue'].shift(1)
df = df.dropna().reset_index(drop=True)
# ---------------------------
# 3. Create the Multivariate Observation Matrix X
# ---------------------------
# We include revenue (as the first feature) plus extra engineered features.
feature_cols = ['revenue', 'sin_day', 'cos_day', 'is_weekend', 'lag1']
X = df[feature_cols].values

# (Optional) Scale features so they are on comparable scales.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ---------------------------
# 4. Fit the Multivariate HMM
# ---------------------------
# Set number of hidden states (e.g., 3).
n_components = 3
hmm_model = GaussianHMM(n_components=n_components, covariance_type="diag", 
                        n_iter=1000, random_state=2)
hmm_model.fit(X_scaled)

# For interpretability, print the transition matrix and the emission means.
print("Transition Matrix (rounded):")
print(np.round(hmm_model.transmat_, 3))

# The emission means are vectors corresponding to each feature.
# We assume the first feature (index 0) is revenue.
print("\nEmission Means (first feature = revenue) before inverse-scaling:")
for i in range(n_components):
    print(f"State {i}: {np.round(hmm_model.means_[i, 0], 3)}")

# To convert the revenue means back to the original scale,
# we invert the scaling transformation for the revenue column.
# The scaler's mean_ and scale_ arrays correspond to the order in feature_cols.
revenue_mean_original = scaler.mean_[0]
revenue_scale = scaler.scale_[0]

print("\nEmission Means (for revenue) in original scale:")
for i in range(n_components):
    # Inverse transform: original_revenue = (scaled_revenue * scale) + mean
    orig_mean = hmm_model.means_[i, 0] * revenue_scale + revenue_mean_original
    print(f"State {i}: {np.round(orig_mean, 3)}")

# ---------------------------
# 5. Rolling 28-Day Forecast Using the HMM
# ---------------------------
# We'll use the posterior probabilities from the last observation as the starting state distribution.
posteriors = hmm_model.predict_proba(X_scaled)
p_current = posteriors[-1]  # state probability distribution for last observed day

forecast_days = 28
forecast_revenue = []
forecast_dates = []
last_date = df['date'].iloc[-forecast_days]

# For a rolling forecast, we update the state probabilities using the transition matrix.
# Since we only forecast revenue, we use the revenue component (first dimension) of the emission means.
# (Note: This approach ignores the fact that future exogenous features are known; it is a first approximation.)
for i in range(forecast_days):
    # Update state distribution:
    p_next = np.dot(p_current, hmm_model.transmat_)
    # Expected forecasted revenue in scaled space (first feature):
    revenue_forecast_scaled = np.dot(p_next, hmm_model.means_[:, 0])
    # Inverse-scale to obtain revenue in original units:
    revenue_forecast = revenue_forecast_scaled * revenue_scale + revenue_mean_original
    forecast_revenue.append(revenue_forecast)
    
    # For the next iteration, we update p_current.
    p_current = p_next
    forecast_dates.append(last_date + timedelta(days=i+1))

# ---------------------------
# 6. Plotting the Historical Data and Forecast
# ---------------------------
plt.figure(figsize=(12, 6))
plt.plot(df['date'][-50:], df['revenue'][-50:], label='Historical Revenue', color='blue')
plt.plot(forecast_dates, forecast_revenue, marker='o', linestyle='--', 
         color='red', label='28-Day Revenue Forecast')
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.title("Rolling 28-Day Forecast for Revenue Using Multivariate HMM")
plt.legend()
plt.grid(True)
plt.show()

# Optionally, display the forecast in a DataFrame.
forecast_df = pd.DataFrame({'date': forecast_dates, 'forecast_revenue': forecast_revenue})
print("\n28-Day Forecast:")
print(forecast_df)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.regime_switching.markov_regression import MarkovRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
from datetime import timedelta

# --- 1. Load Data and Create Features ---
df = pd.read_csv("df.csv")  # your CSV file with 'date' and 'revenue'
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date').reset_index(drop=True).fillna(0)

# Create a weekend indicator as an exogenous variable (can add more exog features)
df['dayofweek'] = df['date'].dt.dayofweek
df['is_weekend'] = df['dayofweek'].isin([5,6]).astype(int)
df['month'] = df['date'].dt.month

# Use sine and cosine transforms for day-of-week to capture cyclical behavior
df['sin_day'] = np.sin(2 * np.pi * df['dayofweek'] / 7)
df['cos_day'] = np.cos(2 * np.pi * df['dayofweek'] / 7)

# Define the endogenous (dependent) variable and the exogenous regressors.
endog = df['revenue']
exog_cols = ['sin_day', 'cos_day', 'month','is_weekend']
exog = df[exog_cols]

# --- 2. Train/Test Split ---
# Reserve the last 28 days for forecasting.
train_size = len(df) - 28
endog_train = endog.iloc[:train_size]
endog_test  = endog.iloc[train_size:]
exog_train  = exog.iloc[:train_size]
exog_test   = exog.iloc[train_size:]
dates_train = df['date'].iloc[:train_size]
dates_test  = df['date'].iloc[train_size:]

# --- 3. Fit the Markov Regression Model ---
# Here we use 3 regimes, include a constant term ('c'), and allow variance to switch.
model = MarkovRegression(endog_train,
                         k_regimes=2,
                         exog=exog_train,
                         trend='c',
                         switching_variance=False)

# Fit the model.
res = model.fit(maxiter=200, em_iter=20)
print(res.summary())

# Smoothed probabilities of each regime over the in-sample period
in_sample_probs = res.smoothed_marginal_probabilities
# shape: (k_regimes, nobs)


# --- 4. In-Sample Prediction ---
# The predict method requires the parameter vector. Use res.params.
in_sample_pred = res.predict()
df_train = pd.DataFrame({
    'date': dates_train,
    'revenue': endog_train,
    'fitted': in_sample_pred
})
plt.figure(figsize=(12, 6))
plt.plot(dates_train, endog_train, label='Actual Revenue', color='blue')
plt.plot(dates_train, in_sample_pred, label='In-Sample Fitted', color='red', linestyle='--')
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.title("Markov Regression In-Sample Fit")
plt.legend()
plt.grid(True)
plt.show()
```

```python
k_regimes = 2
# The transition matrix is (k_regimes x k_regimes), giving P(next_state | current_state).
P = res.regime_transition

# Suppose we have n_exog = len(exog_cols). For switching_exog=True,
# each regime has (n_exog + 1) parameters: an intercept + one slope per exog feature.
n_exog = len(exog_cols)
n_params = n_exog + 1  # intercept + 4 exog
# The first (2*n_params) parameters in results.params are the regime-specific regression parameters.
params = res.params[:2 * n_params]
# Reshape them into a (2, 5) matrix where each row corresponds to one regime.
param_matrix = params.values.reshape(2, n_params)
# Now, for regime r, the linear predictor is:
#    y_hat_r = param_matrix[r, 0] + param_matrix[r,1]*sin_day + param_matrix[r,2]*cos_day + param_matrix[r,3]*month + param_matrix[r,4]*is_weekend


print("Transition matrix:\n", np.round(P, 3))
print("\nRegime-specific regression parameters:")
for i in range(k_regimes):
    print(f"Regime {i}:", param_matrix[i])
```

```python
# -------------------------------
# 3. Generate Future Exogenous Features for 28 Days
# -------------------------------
forecast_days = 28
last_date = df['date'].iloc[-29]
future_dates = [last_date + timedelta(days=i+1) for i in range(forecast_days)]
future_df = pd.DataFrame({'date': future_dates})

# Compute day-of-week (Monday=0, Sunday=6)
future_df['dayofweek'] = future_df['date'].dt.dayofweek
# Compute sine and cosine transforms for day-of-week (7-day cycle)
future_df['sin_day'] = np.sin(2 * np.pi * future_df['dayofweek'] / 7)
future_df['cos_day'] = np.cos(2 * np.pi * future_df['dayofweek'] / 7)
# Use month as numeric (1-12)
future_df['month'] = future_df['date'].dt.month
# Compute weekend indicator: 1 if dayofweek is Saturday (5) or Sunday (6)
future_df['is_weekend'] = future_df['dayofweek'].isin([5, 6]).astype(int)

# Ensure the exogenous features are in the same order as used in the model:
# Here: sin_day, cos_day, month, is_weekend.
forecast_exog = future_df[['sin_day', 'cos_day', 'month', 'is_weekend']].values

# -------------------------------
# 4. Manual Rolling Forecast
# -------------------------------
forecast_revenue = []
# Copy p_current to update it iteratively.
p_current_forecast = p_current.copy()

for i in range(forecast_days):
    # Get exogenous vector for forecast day i: shape (4,)
    x = forecast_exog[i]
    # Build the design vector with an intercept: [1, sin_day, cos_day, month, is_weekend]
    design_vector = np.concatenate(([1], x))
    
    # Compute predicted revenue for each regime
    regime_preds = np.array([np.dot(param_matrix[r], design_vector) for r in range(2)])
    
    # Update regime probability vector: p_next = p_current * P
    p_next = np.dot(p_current_forecast, P)
    p_next = p_next.squeeze()
    # Compute the expected revenue for day i as the weighted sum:
    y_hat = np.dot(p_next, regime_preds)
    
    forecast_revenue.append(y_hat)
    
    # Update p_current for the next iteration
    p_current_forecast = p_next

# -------------------------------
# 5. Visualize the Forecast
# -------------------------------
future_df['forecast_revenue'] = forecast_revenue

plt.figure(figsize=(12, 6))
plt.plot(dates_test, endog_test, label='Historical Revenue', color='blue')
plt.plot(future_df['date'], future_df['forecast_revenue'], marker='o', linestyle='--', 
         color='red', label='28-Day Forecast')
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.title("28-Day Out-of-Sample Forecast using 2-Regime Markov Switching Model")
plt.legend()
plt.grid(True)
plt.show()

# Optionally, print the forecast table
print(future_df[['date', 'forecast_revenue']])
```
