## Import data

```python
calendar = pd.read_csv("calendar.csv")
sales_train_validation = pd.read_csv("sales_train_validation.csv")
sell_prices = pd.read_csv("sell_prices.csv")
```

```python
sales_train_validation_ca_1 = sales_train_validation[
    sales_train_validation["store_id"] == "CA_1"
]
```

```python
sales_train_validation_ca_1
```

```python
sell_prices_ca_1 = sell_prices[sell_prices["store_id"] == "CA_1"]
```

```python
sell_prices_ca_1
```

```python
melted_sales_train_validation_ca_1 = sales_train_validation_ca_1.melt(
    id_vars=["id", "item_id", "dept_id", "cat_id", "store_id", "state_id"],
    var_name="d",
    value_name="amount",
)
```

```python
melted_sales_train_validation_ca_1
```

```python
merged_sales_train_validation_ca_1 = calendar.merge(
    melted_sales_train_validation_ca_1, on="d", how="inner"
)
```

```python
merged_sales_train_validation_ca_1
```

```python
ca_1 = merged_sales_train_validation_ca_1.merge(
    sell_prices_ca_1, on=["store_id", "item_id", "wm_yr_wk"], how="left"
)
```

```python
ca_1
```

```python
ca_1_sell_price_filled = ca_1["sell_price"].fillna(0.0)
```

```python
ca_1["sell_price"] = ca_1_sell_price_filled
```

```python
ca_1["revenue"] = (ca_1["sell_price"] * ca_1["amount"]).round(5)
```

```python
ca_1
```

```python
ca_1_revenue = ca_1.groupby(
    by=[
        "d",
        "date",
        "wm_yr_wk",
        "store_id",
    ]
)["revenue"].sum().reset_index()
```
